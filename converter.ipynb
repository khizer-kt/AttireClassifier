```python
from PIL import Image
import os

def convert_images_to_jpg_and_delete(input_folder):
    # List all files in the input folder
    files = os.listdir(input_folder)

    for file_name in files:
        # Check if the file is an image (you can customize this condition)
        if file_name.lower().endswith(('.png', '.jpeg', '.bmp', '.gif', '.tiff')):
            input_path = os.path.join(input_folder, file_name)

            # Open the image
            img = Image.open(input_path)

            # Convert to RGB if the image has a palette-based mode or RGBA mode
            if img.mode in ['P', 'RGBA']:
                img = img.convert('RGB')

            # Create the output path with the same filename but with a .jpg extension
            output_path = os.path.join(input_folder, os.path.splitext(file_name)[0] + '.jpg')

            # Save the image as JPG
            img.save(output_path, 'JPEG')

            print(f"Converted: {input_path} to {output_path}")

            # Delete the original non-JPG image
            os.remove(input_path)
            print(f"Deleted: {input_path}")

if __name__ == "__main__":
    input_folder = "data2/shalwarKameez"

    convert_images_to_jpg_and_delete(input_folder)
```
